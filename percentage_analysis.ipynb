```
import json
import os
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
files = os.listdir("/content/drive/My Drive/Sentiment analysis")
dates=['2020-03-31','2020-04-16','2020-04-17','2020-04-18','2020-04-19','2020-04-20','2020-04-21','2020-04-22','2020-04-23','2020-04-24','2020-04-25','2020-04-26','2020-04-27','2020-04-29','2020-04-30']
column=['Date','Positive count','Neutral count','Negative count','Total count']
```

```
data_total=[]
for date in dates:
  data=[]
  count=0
  c=[x for x in files if date in x]
  c.sort(reverse=True)
  data.append(date)
  for filename in c:
    with open(filename) as json_file:
      f = json.load(json_file)
    data.append(len(f['tweets']))
    count+=len(f['tweets'])
  data.append(count)
  data_total.append(data)
df = pd.DataFrame(data_total,columns=column)
```

```
df.to_json(r'/content/drive/My Drive/Colab Notebooks/percentage/percentage_analysis.json')
```
